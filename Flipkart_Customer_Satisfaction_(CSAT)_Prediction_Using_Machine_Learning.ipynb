# **Project Summary -**

Flipkart is one of India’s largest e-commerce platforms, serving millions of customers daily across diverse product categories and service channels. In such a highly competitive environment, customer satisfaction (CSAT) has become a critical metric for measuring service quality, customer loyalty, and long-term business success. This project focuses on building a data-driven machine learning solution to analyze customer support interactions and accurately predict CSAT scores, enabling proactive service improvements and better decision-making.

The primary objective of this project is to analyze customer support data and predict customer satisfaction scores (1–5) using historical operational and service-related features. The problem is framed as a classification task, where CSAT acts as the target variable. By predicting customer satisfaction in advance, the business can identify dissatisfied customers early, reduce churn, improve agent performance, and enhance the overall customer experience.

The project begins with Exploratory Data Analysis (EDA) to understand the structure, quality, and patterns within the dataset. The dataset contains multiple variables related to customer issues, support channels, issue categories, agent shifts, agent tenure, resolution time, handling time, item price, and customer remarks. EDA helped uncover important insights such as the distribution of CSAT scores, the presence of missing values, outliers in time-based features, and correlations between service efficiency metrics and customer satisfaction. Various visualizations—including bar charts, box plots, heatmaps, and pair plots—were used to clearly communicate these insights.

Following EDA, data preprocessing and feature engineering were performed to make the dataset suitable for machine learning. Missing values were handled using median imputation for numerical features and “Unknown” labels for categorical variables. Outliers were treated using the IQR method to reduce their impact without losing valuable data. Categorical variables were encoded using label encoding, numerical features were scaled, and log transformations were applied to skewed variables such as resolution time and handling time. Basic textual preprocessing was also applied to customer remarks to ensure future readiness for NLP-based analysis.

After preprocessing, the data was split into training and testing sets using an 80:20 split, ensuring a fair evaluation of model performance. Multiple machine learning models were implemented, including Logistic Regression, Decision Tree, and Random Forest, to compare performance across different algorithms. These models were evaluated using business-relevant metrics such as Accuracy, Precision, Recall, and F1-score, which together provide a balanced understanding of model reliability and real-world impact.
To further enhance performance, hyperparameter optimization using GridSearchCV was applied to all models. Among them, the Random Forest model with optimized hyperparameters consistently outperformed others, delivering the most stable and accurate predictions. Feature importance analysis using Random Forest explainability highlighted key drivers of customer satisfaction, including resolution time, connected handling time, support channel, issue category, agent shift, and agent tenure. These insights are highly valuable from a business perspective, as they directly point to areas where operational improvements can lead to higher customer satisfaction.

Finally, the best-performing model was saved using joblib and successfully reloaded to predict unseen data, confirming its readiness for deployment. This step ensures the solution can be integrated into real-world systems such as dashboards, customer support tools, or automated monitoring pipelines.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Create a copy of the original dataset
df_clean = df.copy()

# Drop unnecessary ID-like columns
cols_to_drop = ['Unique id', 'Order_id']
df_clean.drop(columns=[col for col in cols_to_drop if col in df_clean.columns], inplace=True)

# Convert date columns to datetime format
date_cols = ['Issue_reported at', 'issue_responded', 'order_date_time', 'Survey_response_Date']
for col in date_cols:
    if col in df_clean.columns:
        df_clean[col] = pd.to_datetime(df_clean[col], errors='coerce', dayfirst=True)

# Create resolution time feature (in minutes)
df_clean['resolution_time_mins'] = (
    df_clean['issue_responded'] - df_clean['Issue_reported at']
).dt.total_seconds() / 60

# Remove negative resolution time values
df_clean = df_clean[df_clean['resolution_time_mins'] >= 0]

# Fill missing categorical values with 'Unknown'
cat_cols = df_clean.select_dtypes(include='object').columns
df_clean[cat_cols] = df_clean[cat_cols].fillna('Unknown')

# Fill missing numerical values with median
num_cols = df_clean.select_dtypes(include=['int64', 'float64']).columns
df_clean[num_cols] = df_clean[num_cols].fillna(df_clean[num_cols].median())

print("Dataset is now analysis ready.")
print(df_clean.shape)
````

### What all manipulations have you done and insights you found?

Removed unnecessary ID columns to avoid noise in analysis.

Converted date columns into proper datetime format.

Created a new feature resolution_time_mins to measure service responsiveness.

Handled missing values using median (numerical) and ‘Unknown’ (categorical) imputation.

Removed invalid negative resolution time values.

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Support Channel vs CSAT Score

Null Hypothesis (H₀):
The average CSAT score is the same across different support channels.

Alternate Hypothesis (H₁):
The average CSAT score differs across support channels.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import pearsonr

# Drop missing values
temp = df_clean[['resolution_time_mins', 'CSAT Score']].dropna()

# Perform Pearson correlation test
corr_coef, p_value = pearsonr(
    temp['resolution_time_mins'],
    temp['CSAT Score']
)

print("Correlation Coefficient:", corr_coef)
print("P-value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

Pearson Correlation Test was used to measure the linear relationship between resolution time and CSAT score.

##### Why did you choose the specific statistical test?

Pearson Correlation Test was chosen because it effectively quantifies the linear relationship between resolution time and CSAT score and determines whether the relationship is statistically significant.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Agent Shift vs CSAT Score

Null Hypothesis (H₀):

The mean CSAT score is the same across all agent shifts.

Alternate Hypothesis (H₁):

At least one agent shift has a different mean CSAT score.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import f_oneway

# Group CSAT scores by agent shift
shift_groups = [
    group['CSAT Score'].values
    for _, group in df_clean.groupby('Agent Shift')
]

# Perform One-Way ANOVA test
f_statistic, p_value = f_oneway(*shift_groups)

print("F-statistic:", f_statistic)
print("P-value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

One-Way ANOVA (Analysis of Variance) was used to obtain the p-value.

##### Why did you choose the specific statistical test?


It is suitable because the CSAT Score is numerical and Agent Shift has more than two categories, and we need to compare the mean CSAT scores across multiple groups.

#### 1. Expand Contraction

````python
# Expand Contraction
# Expand Contractions

import re

# Dictionary for common English contractions
contraction_mapping = {
    "can't": "cannot",
    "won't": "will not",
    "don't": "do not",
    "didn't": "did not",
    "isn't": "is not",
    "aren't": "are not",
    "wasn't": "was not",
    "weren't": "were not",
    "hasn't": "has not",
    "haven't": "have not",
    "hadn't": "had not",
    "it's": "it is",
    "i'm": "i am",
    "you're": "you are",
    "they're": "they are",
    "we're": "we are",
    "that's": "that is",
    "there's": "there is",
    "what's": "what is"
}

# Function to expand contractions
def expand_contractions(text):
    for contraction, expanded in contraction_mapping.items():
        text = re.sub(r"\b{}\b".format(contraction), expanded, text)
    return text

# Apply contraction expansion if text column exists
if 'Customer Remarks' in df_clean.columns:
    df_clean['Customer Remarks'] = df_clean['Customer Remarks'].astype(str).str.lower()
    df_clean['Customer Remarks'] = df_clean['Customer Remarks'].apply(expand_contractions)

print("Contraction expansion completed.")
````

````python
# ML Model - 3 : Random Forest with GridSearchCV
# Fit the Algorithm & Predict on the Model

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV


# Step 1: Define base model

rf_model = RandomForestClassifier(
    random_state=42,
    n_jobs=-1
)

# Step 2: Define hyperparameter grid

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

# Step 3: Apply GridSearchCV

grid_search_rf = GridSearchCV(
    estimator=rf_model,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

# Fit the model
grid_search_rf.fit(X_train, y_train)

# Step 4: Best model & prediction

best_rf_model = grid_search_rf.best_estimator_

# Predict on test data
y_pred_rf_opt = best_rf_model.predict(X_test)

print("Best Parameters:", grid_search_rf.best_params_)
print("Random Forest with GridSearchCV trained and predictions generated successfully.")
````
